```python
import pandas as pd
from unidecode import unidecode

liste_prenoms_source = pd.read_csv("prenoms-2023-dpt.tmp.csv", sep=";")
```

```python
df_clean = liste_prenoms_source.copy()

df_clean = df_clean[(df_clean["preusuel"].str.lower() != "_PRENOMS_RARES".lower()) & (df_clean["annais"] != "XXXX")]
# df_clean["preusuel"] = df_clean["preusuel"].apply(lambda x: unidecode(x, "utf-8"))
df_clean["preusuel"] = df_clean["preusuel"].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
df_clean['annais'] = df_clean['annais'].astype('int')
```

```python
# Quel est le prénom masculin et féminin le plus populaire de l'année 1995 dans la France entière (Territoire d'Outre-Mer + Metropole)
annee = 1992
naissances_1995 = (
    df_clean[df_clean["annais"] == annee].groupby(['preusuel', "sexe"], as_index=False)["nombre"].sum()
)

top_naissances = (naissances_1995.groupby(['sexe'], as_index=False).apply(lambda grp: grp.nlargest(1, "nombre")))
top_naissances.reset_index(drop=True, inplace=True)
display(top_naissances)
total_naissances = df_clean[df_clean["annais"] == annee].groupby(["sexe"], as_index=False)["nombre"].sum()
display(total_naissances)

display(top_naissances[(top_naissances["sexe"] == 1)]["nombre"] / total_naissances[(total_naissances["sexe"] == 1)]["nombre"])
# part_garcons = 
```
