```python
import pandas as pd

dictionnaire = {
    'marque': ["BMW", "Volvo", "Ford", "Toyota"],
    'gamme': ["luxe", "haut de gamme", "compacte", "citadine"],
    'vitesse_max': [210, 160, 120, 130],
    'nbre_sieges': [5, 7, 4, 5],
    'longueur': [3000, 2750, 2000, 2500],
    'annee_sortie': [2014, 2020, 2009, 2015],
    'motorisation': ["essence", "essence", "bio-ethanol", "diesel"],
    'prix €' : [65000, 56750, 12500, 32000]
}

mon_df = pd.DataFrame(dictionnaire)
display(mon_df)
```

```python
# Question 1
length_sup_1000 = mon_df[mon_df["longueur"] > 1000]
display(length_sup_1000)

# Question 2
max_speed_160 = mon_df[mon_df["vitesse_max"] <= 160]
display(max_speed_160)

# Question 3
seats_5 = mon_df[(mon_df["nbre_sieges"] > 5) & (mon_df["longueur"] < 2000)]
display(seats_5)

# Question 4
compact = mon_df[(mon_df["gamme"] == "compacte")][["gamme"]]
display(compact)

# Question 5
compact = mon_df[["prix €", "motorisation"]]
display(compact)
```

```python
import pandas as pd
import numpy as np


# Pour avoir un peu plus de données, nous allons charger un fichier csv, le fichier "liste_voitures.csv", 
# ça sera plus simple pour la lisibilité
# Les lignes pour importer ce fichier seront expliquée dans la partie suivante
# Les clés restent
voitures_df = pd.read_csv("../ressources/pandas-et-numpy/liste_voitures.csv", sep=",", encoding="utf-8").set_index('modele')


# Affiche toutes les clés de notre DataFrame
display(voitures_df.columns.values.tolist())
voitures_df.dtypes
```

```python
# Contient tous les véhicules dont la longueur est supérieure à 1 000
length_sup_1000 = voitures_df[voitures_df["longueur (mm)"] > 1000]
display(length_sup_1000)
```

```python
# Contient tous les véhicules dont la vitesse max est inférieure ou égale à 160
max_speed_160 = voitures_df[voitures_df["vitesse_max"] <= 160]
display(max_speed_160)
```

```python
# Contient tous les véhicules dont le nombre de sièges est strictement supérieur à 5 et dont la longueur est inférieure à 1 500
seats_5 = voitures_df[(voitures_df["nbre_sieges"] > 5) & (voitures_df["marque"] == "Volvo")]
display(seats_5)
```

```python
# Contient uniquement les véhicules de la gamme "compacte" et les Series "marque" et "longueur"
compact = voitures_df[(voitures_df["gamme"] == "compacte")][["gamme", "longueur (mm)"]]
display(compact)
```

```python
# Contient tous les véhicules avec une nouvelle Serie (colonne) appelée "vitesse_max m/h" et comme valeur la vitesse_max de chaque véhicule en miles/heure
miles = voitures_df.copy()
miles["vitesse_max m/h"] = np.floor(voitures_df["vitesse_max km/h"] * 0.621371).astype('int')
display(miles)

annee_avec_plus_sorties = voitures_df["annee_sortie"].mode()
display(annee_avec_plus_sorties)
```

```python
# Contient la vitesse moyenne (en km/h) des véhicules par gamme
vit_moyenne = voitures_df.groupby(['gamme'])["vitesse_max km/h"].mean().reset_index()
vit_moyenne
```

```python
# Contient le prix moyen des véhicules par année de sortie
mean_price_year = voitures_df.groupby(['annee_sortie'])["prix €"].median().reset_index().astype(int) 
mean_price_year
```

```python
voitures_df["decade_sortie"] = voitures_df["annee_sortie"] // 10 * 10

# On affiche la colonne "marque" et la nouvelle colonne "decade_sortie"
# Par exemple 2014 deviendra 2010 dans la colonne "decade_sortie"
display(voitures_df[["marque", "decade_sortie"]])
```
